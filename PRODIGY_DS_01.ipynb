```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
ages = np.random.randint(18, 80, size=100)

plt.figure(figsize=(8, 6))
plt.hist(ages, bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution in a Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

genders = ['Male', 'Female', 'Other']
counts = [40, 50, 10]

plt.figure(figsize=(8, 6))
plt.bar(genders, counts, color=['blue', 'pink', 'gray'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

colors = ['Red', 'Blue', 'Green', 'Yellow']
counts = [25, 40, 30, 15]

plt.figure(figsize=(8, 6))
plt.bar(colors, counts, color=['red', 'blue', 'green', 'yellow'])
plt.title('Favorite Colors Distribution')
plt.xlabel('Color')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

genders = ['Male', 'Female', 'Other']
gender_counts = [50, 65, 20]

np.random.seed(2)
ages = np.random.randint(20, 70, size=150)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

ax1.bar(genders, gender_counts, color=['blue', 'pink', 'gray'])
ax1.set_title('Gender Distribution')
ax1.set_xlabel('Gender')
ax1.set_ylabel('Count')
ax1.grid(axis='y', linestyle='--', alpha=0.7)


ax2.hist(ages, bins=15, color='lightblue', edgecolor='black')
ax2.set_title('Age Distribution')
ax2.set_xlabel('Age')
ax2.set_ylabel('Frequency')
ax2.grid(True)

plt.tight_layout()
plt.show()
```
